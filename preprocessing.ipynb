```python
import pandas as pd

new_df = pd.read_excel("/content/propaganda_translator_datasheet.xlsx")

new_df["propaganda_label"] = new_df["propaganda_label"].map({
    -1: 0,
     1: 1
})
```

```python
new_df.shape
```

```python
new_df = new_df[["article_text_en", "propaganda_label"]]

new_df = new_df.rename(columns={
    "article_text_en": "text",
    "propaganda_label": "label"
})
```

```python
new_df.to_csv("new_dataset_clean.csv", index=False)
```

```python
old_df = pd.read_csv("/content/propaganda_dataset_for_bert.csv") #importing old dataset
```

```python
old_df = old_df[["text","label"]]
```

```python
combined_df = pd.concat([old_df, new_df], ignore_index=True) #merging the datasets
```

```python
combined_df = combined_df.sample(frac=1, random_state=42)  #shuffling
```

```python
 combined_df.to_csv("combined_dataset.csv", index=False)  #saving the combined csv
```

```python
import re

def clean_text(text):

    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text)

    return text
```

```python
df["text"] = df["text"].apply(clean_text)
```
